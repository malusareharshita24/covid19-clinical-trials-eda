## ðŸ“Š Visualization: Study Type Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Study Type', data=df, order=df['Study Type'].value_counts().index)
plt.title('Distribution of Study Types')
plt.xlabel('Count')
plt.ylabel('Study Type')
plt.tight_layout()
plt.show()
```

## ðŸ¦  Filter COVID-Related Trials

```python
covid_df = df[df['Title'].str.contains('COVID', na=False, case=False)]
covid_df.shape[0]
```

## ðŸ•’ Trials Over Time (Start Year Analysis)

```python
if 'Start Date' in df.columns:
    df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
    df['Year'] = df['Start Date'].dt.year
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x='Year', order=sorted(df['Year'].dropna().unique()))
    plt.title('Trials by Start Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Trials')
    plt.tight_layout()
    plt.show()
```
